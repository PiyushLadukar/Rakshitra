```python
import pandas as pd

df = pd.read_csv("../data/crime_dataset_india.csv")

print(df.shape)
print(df.columns)
print(df.head())
print(df.info())
```

```python
df = df[['Date of Occurrence', 'City', 'Crime Description']]
```

```python
df['Date of Occurrence'] = pd.to_datetime(
    df['Date of Occurrence'],
    format='%d-%m-%Y %H:%M',
    errors='coerce'
)
```

```python
df = df.dropna(subset=['Date of Occurrence'])
```

```python
df['date'] = df['Date of Occurrence'].dt.date
```

```python
df = df[df['City'] == 'Pune']
```

```python
df_grouped = df.groupby('date').size().reset_index(name='crime_count')
```

```python
df_grouped = df_grouped.sort_values('date')
```

```python
print(df_grouped.head())
```

```python
df_grouped['date'] = pd.to_datetime(df_grouped['date'])
```

```python
df_grouped = df_grouped.set_index('date')
```

```python
df_grouped = df_grouped.asfreq('D', fill_value=0)
```

```python
df_grouped = df_grouped.reset_index()
```
